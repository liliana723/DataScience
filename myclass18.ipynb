REGRESSION LINEAL

```python
# establecemos los valores a las graficas
X = df_2.drop("median_house_value", axis=1)
Y = df_2["median_house_value"]

from sklearn.model_selection import train_test_split
X_entrena, X_prueba, Y_entrena, Y_prueba = train_test_split(X, Y, test_size=.2)
```

```python
# muestra la cantidad de registros y columnas
X_entrena.shape
```

```python
from sklearn.linear_model import LinearRegression

modelo=LinearRegression()
modelo.fit(X_entrena, Y_entrena)
```

```python
prediccion = modelo.predict(X_prueba)
print(prediccion)
```

```python
comparativa = {"Prediccion": prediccion, "Valor Real": Y_prueba}
pd.DataFrame(comparativa)
```

```python
print(modelo.score(X_prueba, Y_prueba))
print(modelo.score(X_entrena, Y_entrena))
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_prueba, prediccion)
mse
```

```python
# raiz cuadrada del error medio (RMSE)
import numpy as np
rmse = np.sqrt(mse)
rmse
```

```python
# Evaluar el modelo con validacin cruzada (mtrica: error cuadratico)
from sklearn.model_selection import cross_val_score
scores = cross_val_score(modelo, X, Y, cv=10, scoring='neg_mean_squared_error')
scores
```

```python
rmse_scores = np.sqrt(-scores)
rmse_scores
```

```python
def mostrar_scores(scores):
  print("Puntajes de RMSE en cada fold:", scores) # MOSTRAR LA RAIZCUADRADA DEL ERROR CUADRATICO MEDIO
  print("Promedio de MSE:", np.mean(scores)) # HAYAR PROMEDIO
  print("Desviacin estndar de MSE:", np.std(scores)) # HAYAR DESVIACION ESTANDAR

mostrar_scores(rmse_scores)
```
